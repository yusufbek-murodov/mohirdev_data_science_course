```python
import pandas as pd
import numpy as np
```

```python
data = {'Yil':[2021, 2020, 2019, 2018, 2017, 2016, 2015, 2010],
       'Aholi soni':[33.9, 33.5, 32.9, 32.5, 31.9, 31.4, 30.9, 28.5],
       'Temp':[1.54, 1.48, 1.56, 1.62, 1.65, 1.66, 1.64, 1.53],
       'Zichlik':[79.77, 78.68, 77.53, 76.34, 75.13, 73.91, 72.71, 67.03]}
```

**.reindex** yordamida yangi indekslar qo'shish va turli methodlar yordamida ham to'ldirish mumkin

## DataFramedan elementlarni o'chirish **.drop()**

DataFramedan biror qator yoki ustunni o'chirish uchun qator (ustun) indeksini bilish kifoya.

```python
df
```

```python
df_new = df.drop(10) # 10 indeksini o'chiramiz va yangi indeks yasaymiz
```

```python
df_new
```

```python
df.drop([15, 20]) # 15 va 20 qatorlarni tashlab yuboramiz
```

**.drop()** yordamida nafaqat qatorlarni balki ustunlarni ham o'chirish mumkin. Buning uchun **.drop()** methodining **axis** parametri ishlatiladi.

```python
df.drop('Temp', axis=1) # axis=1 ustunni anglatadi, axis=0 esa qatorni
```

```python
df.drop('Temp', axis='columns') # bu usul ham yuqoridagi natijani beradi
```

Yana bir bor eslatamaiz methodlar asl **DataFrame**ni o'zgartirmaydi

```python
df
```

Lekin buning ham imkoni bir **.drop()** methodlar odatda **inplace** parametri bilan birgalikda keladi odatda **inplace=False** bo'ladi

```python
df.drop('Yil', axis=1, inplace=True)
```
