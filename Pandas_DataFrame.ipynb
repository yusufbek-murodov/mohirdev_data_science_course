# DataFrame ma'lumotlar tuzilmasi

#### **DataFrame** tartibli ustunlardan iborat, ikki o'lchovli (jadval ko'rinishidagi) ma'lumotlar tuzilmasi. Har bir ustun bir turdagi ma'lumotlarni saqlaydi (sonlar, matnlar, Boolen va hokazo.) **DataFrame** ustunlari ham qatorlari ham indekslangan bo'ladi. Umuman olganda **DataFrame**ni bir nechta **Series**dan iborat lug'atlar deb tasavvur qilishimiz mumkin.

```python
import pandas as pd
import numpy as np
from pandas import DataFrame
```

### **DataFrame** yaratishning bir nechta usuli bor va bulardan eng osoni qiymatlari bir hil uzunlikdagi ro'yxatdan iborat lug'at orqali yaratish.

```python
data = {'Yil': [2021, 2020, 2019, 2018, 2017, 2016, 2015, 2010],
        'Aholi soni': [33.9, 33.5, 32.9, 32.5, 31.9, 31.4, 30.9, 28.5],
        'Temp': [1.54, 1.48, 1.56, 1.62, 1.65, 1.66, 1.64, 1.53]}

df = DataFrame(data)
```

```python
df
```

```python
df.head()
```

Odatda katta **DataFrame**lar bilan ishlashda **.head()** metodi yordamida jadvalning 5 ta qatorini ko'rishimiz mumkin.

```python
df.head()
```

Lug'atdan **DataFrame** yaratishda har bir qator indekslanadi (0, 1, 2...), ustunlar ketma-ketligi esa lug'atdagi ketma-ketlikka bog'liq bo'ladi. Ustunlar ketma-ketligini berishni istasak **DataFrame** yaratishda **columns** parametrini ham ko'rsatib ketishimiz mumkin.

```python
df = DataFrame(data, columns=['Yil', 'Temp', 'Aholi soni'])
```

```python
df
```

```python
# Agar columns ro'yhtatida lug'atda mavjud bo'lmagan ustin bo'lsa, shu ustunga bo'sh qiymatlar qo'yilib ketiladi.
```

```python
df = DataFrame(data, columns=['Yil', 'Temp', 'Aholi soni', 'YAIM'])
```

```python
df.head()
```

```python
# Huddi Series kabi DataFrame indekslariga ham o'zimiz qiymat berishimiz mumkin:
df = DataFrame(data, columns=['Yil', 'Temp', 'Aholi soni', 'YAIM'], index=[21, 20, 19, 18, 17, 16, 15, 10])
```

```python
df
```

**DataFrame**ning istalgan ustunini **Series** ko'rinishida ajratib olishimiz mumkin. Buning ikki usuli bor:
 1. Ustun nomi atribut sifatida

```python
df.Temp
```

2. Lug'at kabi ajratish

```python
df['Aholi soni']
```

1-Usul faqatgina ustun nomi bir so'zdan iborat, Python o'zgaruvchi talablariga javob bergan holda ishlaydi.

Ahamiyat bering, ajratib olingan ustunlar indeksi va ustun nomi saqlanib qoladi

Yuqoridagi usul bilan biz ustundagi barcha elementlarga qiymat berishimiz mumkin

```python
df.YAIM = 1500
df
```

**Series** yordamida biror ustunning ma'lum bir qatoriga o'zgartirish kiritish mumkin. Bunda qatorlarning indeksi berilishi shart.

```python
val = pd.Series([1750, 1560, 1270], index=[21, 18, 10])
```

```python
df.YAIM = val
```

```python
df
```

Shuningdek ustun qiymatlarini o'zgartirish uchun shu ustun qatorlari soniga teng ro'yxatdan foydalanishimiz mumkin.

```python
YAIM = [1700, 1600, 1500, 1400, 1550, 1480, 1630, 1200]
```

```python
df['YAIM'] = YAIM
```

```python
df
```

Agar qiymat berishda mavjud bo'lmagan ustun nomi berilsa DataFrameda yangi ustun qo'shiladi.

Quyida mizolda biz avval yangi **1500+** degan ustun yaratdik keyin esa bu ustun qiymatlarini **df.YAIM > 1500** sharti yordamida to'lirdik (ya'ni yalpi ichki mahsulot **1500** dan katta bo'lsa True bo'ladi)

```python
df['1500+'] = df.YAIM >= 1500
```

```python
df
```

Huddi lug'atlardagi kabi ustunni **del** operatori yordamida o'chib tashlash mumkin

```python
del df['1500+']
```

```python
df
```
